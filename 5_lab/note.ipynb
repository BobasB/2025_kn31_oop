# Розглядаємо принцип Інкапсуляції в об'єктно-орієнтованому програмуванні
- створимо програму яка буде визначати наскільки ми знудились на лекції

```python
class CheckStudentBoredOnLecture:
    def __init__(self, name, boredom_level):
        self.__boredom_level = boredom_level  # Приватний атрибут
        self.name = name  # Публічний атрибут
    
    def __words_sponen_on_lecture(self):
        import random
        words = random.randint(0, 100)
        return words
    
    def __check_is_bored(self):
        words_spoken = self.__words_sponen_on_lecture()
        if words_spoken < self.__boredom_level:
            return True
        else:
            return False
    
    def get_boredom_status(self):
        if self.__check_is_bored():
            return f"{self.name} дуже нудьгує на лекції."
        else:
            return f"{self.name} уважно слухає на лекції."
        
    
student = CheckStudentBoredOnLecture("Максим", 20)
student.get_boredom_status()
```

# Напишемо приклад Наслідування в Python
- ми хочемо змоделювати систему забезпечення електроенергії
- створити клас наслудування який генерує електроенергію з Сонця або вітру або палива
- створити гібридний клас який буде поєнувати це невідомі джерела енергії

```python
class ElectricPowerDevice:
    def __init__(self, power_capacity):
        self.power_capacity = power_capacity
    
    @property
    def info(self):
        return f"Потужність пристрою: {self.power_capacity} Вт"
    
    def charge_device(self):
        return "Пристрій заряджається."
    
    def calculate_total_power(self, additional_power):
        return (self.power_capacity + additional_power) * 0.95  # Враховуємо 5% втрат
    
    def produce_electricity(self):
        return f"Пристрій виробляє {self.power_capacity} Вт електроенергії."

class SolarPanelDevice(ElectricPowerDevice):
    def __init__(self, power_capacity, panel_area):
        super().__init__(power_capacity)
        self.panel_area = panel_area
    
    def __calculate_power_from_panel(self):
        return self.panel_area * 150 # Припустимо, що кожен м² панелі виробляє 150 Вт
    
    def __calculate_total_power(self):
        return self.power_capacity + self.__calculate_power_from_panel()
    
    def produce_electricity(self):
        return f"Сонячна електростанція буде виробляти в загальному: {self.__calculate_total_power()} Вт."

    @property
    def total_power_capacity(self):
        return self.__calculate_total_power()

class HybridPowerDevice(SolarPanelDevice):
    def __init__(self, power_capacity, panel_area, wind_turbine_power):
        super().__init__(power_capacity, panel_area)
        self.wind_turbine_power = wind_turbine_power
    
    def produce_electricity(self):
        total_power = self.total_power_capacity + self.wind_turbine_power
        return f"Гібридна електростанція буде виробляти в загальному: {total_power} Вт."
    
class Generator(ElectricPowerDevice):
    def __init__(self, power_capacity, fuel_capacity):
        super().__init__(power_capacity)
        self.fuel_capacity = fuel_capacity
    
    def __calculate_power_from_fuel(self):
        return self.fuel_capacity * 100 # Припустимо, що кожен літр палива виробляє 100 Вт

    def produce_electricity(self):
        return f"Генератор в загальному виробляє: {self.calculate_total_power(self.__calculate_power_from_fuel())} Вт."

class TeslaPowerDevice:
    def __init__(self, power_capacity, ether_capacity):
        self.power_capacity = power_capacity
        self.ether_capacity = ether_capacity
    
    @property
    def info(self):
        return f"Потужність пристрою: {self.power_capacity} Вт, Ємність ефіру: {self.ether_capacity} Дж"
    
    @property
    def total_power_capacity(self):
        return self.power_capacity + self.ether_capacity * 0.1  # Припустимо, що ефір додає 10% до потужності

    def produce_electricity(self):
        return f"Пристрій виробляє {self.total_power_capacity} Вт електроенергії."

# Множинне наслідування
class TeslaHybridPowerDevice(HybridPowerDevice, TeslaPowerDevice):
    def __init__(self, power_capacity, panel_area, wind_turbine_power, ether_capacity):
        HybridPowerDevice.__init__(self, power_capacity, panel_area, wind_turbine_power)
        TeslaPowerDevice.__init__(self, power_capacity, ether_capacity)
    
    def produce_electricity(self):
        total_power = self.total_power_capacity + self.wind_turbine_power
        return f"Тесла гібридна електростанція буде виробляти в загальному: {total_power} Вт."
    
    
D = ElectricPowerDevice(500)
print(D.info)
print(D.charge_device())
print(D.produce_electricity())

S = SolarPanelDevice(500, 5)
print(S.produce_electricity())
print(S.charge_device())
print(S.info)

H = HybridPowerDevice(500, 5, 300)
print(H.produce_electricity())
print(H.charge_device())
print(H.info)

G = Generator(500, 10)
print(G.produce_electricity())
print(G.charge_device())
print(G.info)

T = TeslaHybridPowerDevice(500, 5, 300, 200)
print(T.produce_electricity())
print(T.info)
```

# Приклад Поліморфізму в Python
1. реалізація поліморфізму у функціях з різною кількістю аргументів або різними типами даних
2. поліморфізим з наслідуванням, коли нащадок створює класи з однаковими методами, але різною реалізацією
3. поліморфізм через Duck Typing, коли об'єкти різних класів можуть використовуватися взаємозамінно, якщо вони мають однакові методи або атрибути

```python
def device_info():
    return "Це пристрій для виробництва електроенергії."
# тут ми просто переписали функцію з іншим параметром і це не є поліморфізмом
def device_info(device): # це НЕ є поліморфізм
    return f"Це {device} для виробництва електроенергії."

print(device_info("Пристрій"))

# поліморфізм через дефолтні параметри
def device_info(device="пристрій"): # це є поліморфізм
    return f"Це {device} для виробництва електроенергії."

print(device_info(), device_info("сонячна панель"))

# поліморфізм через аргументи змінної довжини або ключові слова
def device_info(*args, **kwargs): # це є поліморфізм
    info = "Це пристрій для виробництва електроенергії."
    if args:
        info += " Аргументи: " + ", ".join(args) + "."
    if kwargs:
        info += " Ключові слова: " + ", ".join(f"{k}={v}" for k, v in kwargs.items()) + "."
    return info

print(device_info())
print(device_info("сонячна панель", "генератор"))
print(device_info(type="гібридний", power="500Вт"))
print(device_info("1", "сонячна панель", type="гібридний", power="500Вт"))
```

```python
class PawerBank:
    def __init__(self, capacity):
        self.capacity = capacity
    
    def charge_device(self, device):
        device.charge()
        return f"Павербанк заряджає {device.name}."
    
    def produce_electricity(self):
        return f"Павербанк виробляє {self.capacity} Вт електроенергії."

class SolarSystem(PawerBank):
    def __init__(self, capacity, panel_area):
        super().__init__(capacity)
        self.panel_area = panel_area
    
    # Перевизначення методу produce_electricity для сонячної системи реалізує принцип поліморфізму
    def produce_electricity(self):
        total_power = self.capacity + self.panel_area * 150
        return f"Сонячна система виробляє {total_power} Вт електроенергії."

O = []
for i in range(3):
    O.append(PawerBank(100 * (i + 1)))
    O.append(SolarSystem(100 * (i + 1), 2 * (i + 1)))

import random
random.shuffle(O)

for P in O:
    print(P.produce_electricity())
```

# Приклад Абстракції в Python
- спробуємо змоделювати роботу з LLM (Large Language Model)
- створимо абстрактний клас який буде визначати основні методи для роботи з LLM
- створимо конкретні класи які реалізують ці методи для різних моделей LLM
- спробуємо також використати typing.Protocol для визначення інтерфейсу роботи з LLM

```python
# Це ще не є приклад абстаркції
class LLMInterface:
    def generate_text(self, prompt: str) -> str:
        pass
    
    def summarize_text(self, text: str) -> str:
        pass

class SimpleLLM(LLMInterface):
    # Це поліморфізи
    def generate_text(self, prompt: str) -> str:
        return f"Generated text based on prompt: {prompt}"

    # тут я не забовязаний реалізовувати другий метод

SL = SimpleLLM()
print(SL.generate_text("Hello, world!"), SL.summarize_text("This is a long text that needs summarization."))
```

```python
from abc import ABC, abstractmethod

class LLMInterfaceAbstraction(ABC):
    @abstractmethod
    def enter_prompt(self, prompt: str) -> None:
        pass
    
    @abstractmethod
    def generate_text(self, prompt: str) -> str:
        pass
    
    @abstractmethod
    def summarize_text(self, text: str) -> str:
        pass

class AdvancedLLM(LLMInterfaceAbstraction):
    # згідно абстракції я зобовязаний реалізувати обидва методи
    def generate_text(self, prompt: str) -> str:
        return f"Advanced generated text based on prompt: {prompt}"
    # якщо реалізую один а інший ні то буде помилка
    def summarize_text(self, text: str) -> str:
        return f"Summarized text: {text[:50]}..."  # Просте скорочення тексту для прикладу
    
    def enter_prompt(self, prompt: str) -> None:
        print(f"Prompt entered: {prompt}")

AL = AdvancedLLM()
print(AL.summarize_text("This is a long text that needs summarization."))
```

```python
from typing import Protocol

class LLMProtocol(Protocol):
    def get_llm_name(self) -> str:
        ...
    
class CustomModel(LLMInterfaceAbstraction):
    def enter_prompt(self, prompt: str, model_name: LLMProtocol) -> None:
        print(f"{model_name.get_llm_name()} received prompt: {prompt}")
    
    def generate_text(self, prompt: str) -> str:
        return f"GPT-4 generated text for prompt: {prompt}"
    
    def summarize_text(self, text: str) -> str:
        return f"GPT-4 summary: {text[:50]}..."

class GPT4Model:
    def get_llm_name(self) -> str:
        return "GPT-4 Model"

class BrokenModel:
    def some_other_method(self) -> str:
        return "This model does not conform to LLMProtocol."

GPT = CustomModel()
GPT.enter_prompt("Поясни абстракцію.", GPT4Model())
#GPT.enter_prompt("Поясни абстракцію.", BrokenModel())  # Це викличе помилку типу
    
```
