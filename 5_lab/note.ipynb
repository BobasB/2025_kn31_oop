# Розглядаємо принцип Інкапсуляції в об'єктно-орієнтованому програмуванні
- створимо програму яка буде визначати наскільки ми знудились на лекції

```python
class CheckStudentBoredOnLecture:
    def __init__(self, name, boredom_level):
        self.__boredom_level = boredom_level  # Приватний атрибут
        self.name = name  # Публічний атрибут
    
    def __words_sponen_on_lecture(self):
        import random
        words = random.randint(0, 100)
        return words
    
    def __check_is_bored(self):
        words_spoken = self.__words_sponen_on_lecture()
        if words_spoken < self.__boredom_level:
            return True
        else:
            return False
    
    def get_boredom_status(self):
        if self.__check_is_bored():
            return f"{self.name} дуже нудьгує на лекції."
        else:
            return f"{self.name} уважно слухає на лекції."
        
    
student = CheckStudentBoredOnLecture("Максим", 20)
student.get_boredom_status()
```

# Напишемо приклад Наслідування в Python
- ми хочемо змоделювати систему забезпечення електроенергії
- створити клас наслудування який генерує електроенергію з Сонця або вітру або палива
- створити гібридний клас який буде поєнувати це невідомі джерела енергії

```python
class ElectricPowerDevice:
    def __init__(self, power_capacity):
        self.power_capacity = power_capacity
    
    @property
    def info(self):
        return f"Потужність пристрою: {self.power_capacity} Вт"
    
    def charge_device(self):
        return "Пристрій заряджається."
    
    def calculate_total_power(self, additional_power):
        return (self.power_capacity + additional_power) * 0.95  # Враховуємо 5% втрат
    
    def produce_electricity(self):
        return f"Пристрій виробляє {self.power_capacity} Вт електроенергії."

class SolarPanelDevice(ElectricPowerDevice):
    def __init__(self, power_capacity, panel_area):
        super().__init__(power_capacity)
        self.panel_area = panel_area
    
    def __calculate_power_from_panel(self):
        return self.panel_area * 150 # Припустимо, що кожен м² панелі виробляє 150 Вт
    
    def __calculate_total_power(self):
        return self.power_capacity + self.__calculate_power_from_panel()
    
    def produce_electricity(self):
        return f"Сонячна електростанція буде виробляти в загальному: {self.__calculate_total_power()} Вт."

    @property
    def total_power_capacity(self):
        return self.__calculate_total_power()

class HybridPowerDevice(SolarPanelDevice):
    def __init__(self, power_capacity, panel_area, wind_turbine_power):
        super().__init__(power_capacity, panel_area)
        self.wind_turbine_power = wind_turbine_power
    
    def produce_electricity(self):
        total_power = self.total_power_capacity + self.wind_turbine_power
        return f"Гібридна електростанція буде виробляти в загальному: {total_power} Вт."
    
class Generator(ElectricPowerDevice):
    def __init__(self, power_capacity, fuel_capacity):
        super().__init__(power_capacity)
        self.fuel_capacity = fuel_capacity
    
    def __calculate_power_from_fuel(self):
        return self.fuel_capacity * 100 # Припустимо, що кожен літр палива виробляє 100 Вт

    def produce_electricity(self):
        return f"Генератор в загальному виробляє: {self.calculate_total_power(self.__calculate_power_from_fuel())} Вт."

class TeslaPowerDevice:
    def __init__(self, power_capacity, ether_capacity):
        self.power_capacity = power_capacity
        self.ether_capacity = ether_capacity
    
    @property
    def info(self):
        return f"Потужність пристрою: {self.power_capacity} Вт, Ємність ефіру: {self.ether_capacity} Дж"
    
    @property
    def total_power_capacity(self):
        return self.power_capacity + self.ether_capacity * 0.1  # Припустимо, що ефір додає 10% до потужності

    def produce_electricity(self):
        return f"Пристрій виробляє {self.total_power_capacity} Вт електроенергії."

# Множинне наслідування
class TeslaHybridPowerDevice(HybridPowerDevice, TeslaPowerDevice):
    def __init__(self, power_capacity, panel_area, wind_turbine_power, ether_capacity):
        HybridPowerDevice.__init__(self, power_capacity, panel_area, wind_turbine_power)
        TeslaPowerDevice.__init__(self, power_capacity, ether_capacity)
    
    def produce_electricity(self):
        total_power = self.total_power_capacity + self.wind_turbine_power
        return f"Тесла гібридна електростанція буде виробляти в загальному: {total_power} Вт."
    
    
D = ElectricPowerDevice(500)
print(D.info)
print(D.charge_device())
print(D.produce_electricity())

S = SolarPanelDevice(500, 5)
print(S.produce_electricity())
print(S.charge_device())
print(S.info)

H = HybridPowerDevice(500, 5, 300)
print(H.produce_electricity())
print(H.charge_device())
print(H.info)

G = Generator(500, 10)
print(G.produce_electricity())
print(G.charge_device())
print(G.info)

T = TeslaHybridPowerDevice(500, 5, 300, 200)
print(T.produce_electricity())
print(T.info)
```
