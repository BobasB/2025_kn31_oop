Створити клас ProductCalculator, який:

```python
class ProductCalculator:
    def __init__(self, numbers: list):
        # Зберігаємо вхідний список чисел
        self.numbers = numbers
        self._product = None

    def calculate_product(self):
        # Обчислюємо добуток усіх елементів списку
        result = 1
        for number in self.numbers:
            result *= number
        self._product = result

    @property
    def product(self):
        # Повертаємо добуток чисел
        if self._product is None:
            self.calculate_product()
        return self._product

    def __str__(self):
        # Вивід інформації про студента
        return "Ім'я студента: Богдан, Група: КН-31"


# Створення декількох об'єктів класу з різними вхідними даними
calculators = [
    ProductCalculator([1, 2, 3, 4]),
    ProductCalculator([2, 5, 10]),
    ProductCalculator([3, 3, 3])
]

for calc in calculators:
    print(calc)
    print(f"Список чисел: {calc.numbers}")
    print(f"Добуток чисел: {calc.product}")
    print("-" * 40)
```
