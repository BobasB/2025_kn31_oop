# Вчимось писати класи на ООП
Система управління бібліотекою

Студенти створюють програму, яка моделює роботу бібліотеки.
Основні класи:

Book (книга: назва, автор, рік, ISBN, статус)
User (користувач: ім'я, ID, список взятих книг)
Library (бібліотека: список книг, список користувачів, методи видачі/повернення книг)

```python
import random
```

```python
from enum import Enum

class BookStatus(Enum):
    AVAILABLE = 'в наявності'
    BORROWED = 'взята'
    RESERVED = 'зарезервована'

class Book:
    def __init__(self, title, author, year, isbn):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.status = BookStatus.AVAILABLE
    
    @property
    def description(self):
        return f"""
                Назва: {self.title}.
                Автор: {self.author}.
                Видання: ({self.year}) - ISBN: {self.isbn}.
                Статус: {self.status.value}.
                """

# Example usage
book_shelf = []
for b in ["Філософський камінь", "Таємна кімната", "В'язень Азкабану", "Кубок вогню", "Орден Фенікса", "Напівкровний принц", "Смертельні реліквії"]:
    book = Book(f"Гаррі Поттер і {b}", "Дж. К. Роулінг", 1997, "978-3-16-148410-0")
    book_shelf.append(book)
    print(book.description)
```

```python
class User:
    def __init__(self, name):
        self.name = name
        self.ID = id(self)
        self.borrowed_books = []
        self.college_name = "ІТ коледж"
    
    @property
    def studet_id_card(self):
        return f"Студент коледжу <<{self.college_name}>>\nІмя: {self.name},\nID: {self.ID}"
    
    def borrow_book(self, book: Book):
        if book.status == BookStatus.AVAILABLE:
            book.status = BookStatus.BORROWED
            self.borrowed_books.append(book)
            return True
        return False
    
    def return_book(self, book: Book):
        if book in self.borrowed_books:
            book.status = BookStatus.AVAILABLE
            self.borrowed_books.remove(book)
            return True
        return False
    
    def start_reading_book(self, book: Book):
        if book in self.borrowed_books:
            return f"{self.name} починає читати описа книги:\n{book.description}"
        return f"{self.name} не має цієї книги"
    
    def think_before_take_a_book(self, book: Book):
        if random.choice([True, False]):
            print( f"{self.name} розглядає книгу:\n{book.description}" )
            return True
        print( f"{self.name} відмовляється від книги:\n{book.description}" )
        return False

Bohdan = User("Богдан")
print(Bohdan.studet_id_card)

Rostuslav = User("Ростислав")
print(Rostuslav.studet_id_card)
```

```python

for day in range(1, 6):
    print(f"\n--- День {day} ---")
    for user in [Bohdan, Rostuslav]:
        book_to_borrow: Book = random.choice(book_shelf)
        if user.think_before_take_a_book(book_to_borrow):
            if user.borrow_book(book_to_borrow):
                print(f"{user.name} успішно взяв книгу:\n{book_to_borrow.description}")
            else:
                print(f"Книга '{book_to_borrow.title}' недоступна для читання.")
        else:
            print(f"{user.name} не взяв книгу '{book_to_borrow.title}'.")
    
        return_random_book: Book = random.choice(user.borrowed_books) if user.borrowed_books else None
        if return_random_book and user.return_book(return_random_book):
            print(f"{user.name} повернув книгу:\n{return_random_book.description}")
```
